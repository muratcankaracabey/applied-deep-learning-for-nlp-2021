```python
import pandas as pd
import seaborn as sns
from langdetect import detect
import datetime
```

```python
comments_df = pd.read_csv('comments_big.csv',engine='python')
print(comments_df.shape)
comments_df.dropna(inplace=True)
print(comments_df.shape)
```

```python
comments_df.head()
```

```python
## Remove Emojis
import demoji
demoji.download_codes()
```

```python
comments_df["clean_comments"] = comments_df["comment"].apply(lambda x: demoji.replace(x,""))
```

```python
## Language Detection
from langdetect import detect

comments_df["lan"] = 0

count = 0
for i in range(0,len(comments_df)):
    temp = comments_df["clean_comments"].iloc[i]
    count += 1
    try:
        comments_df["lan"].iloc[i] = detect(temp)
    except:
        comments_df["lan"].iloc[i] = "no_lan"
```

```python
comments_df = comments_df[comments_df["lan"] == "en"]
```

```python
## Remove Special Characters
regex = r"[^0-9A-Za-z'\t]"
import re
```

```python
comments_df["reg"] = comments_df["clean_comments"].apply(lambda x:re.findall(regex,x))
comments_df["regular_comments"] = comments_df["clean_comments"].apply(lambda x:re.sub(regex,"  ",x))
```

```python
comments_df = comments_df[["video_id","video_title","comment_date","regular_comments","like_count"]]
```

```python
comments_df = comments_df.rename(columns = {"regular_comments":"comments"})
comments_df.head()
```

```python
quantiles_vid = dict()
for vid in comments_df['video_id'].unique():
    try:
        if len(vid) == 11:
            quantiles_vid[vid] = comments_df[(comments_df['video_id'] == vid)].quantile([.3]).values.flatten()
    except:
        print("couldn't find quantiles for vid:",vid)
```

```python
quantiles_vid
```

```python
def find_time_delta(x):
    x = x[2]
    time_delta = datetime.date.today() - datetime.date(int(x[:4]),int(x[5:7]),int(x[8:10]))
    return time_delta.days
```

```python
def find_sentiment(x):
    if x[4] <= quantiles_vid[x[0]][0]:
        return 0
    elif x[4] > quantiles_vid[x[0]][0]:
        return 1
```

```python
comments_df["sentiment"] = comments_df.apply(find_sentiment,axis=1) 
```

```python
comments_df["comment_age"] = comments_df.apply(find_time_delta,axis=1)
```

```python
sns.countplot(comments_df["sentiment"])
```

```python
sns.countplot(comments_df["comment_age"])
```

```python
from matplotlib import pyplot as plt
for i in range(50):
    try:
        print("comment age is ",i)
        sns.countplot(comments_df[comments_df["comment_age"] == i]["sentiment"])
        plt.show()
    except:
        pass
```

```python
final_df = comments_df[(comments_df['comment_age'] > 7) |
                       (comments_df['sentiment'] == 1)].copy()
```
